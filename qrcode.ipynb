```python
import qrcode

# The URL you want to encode
url = "https://sdu-nauryz2026.website.yandexcloud.kz/"

# 1. Configure the QR Code generator
qr = qrcode.QRCode(
    version=1,  # Controls the size of the QR Code (1 is the smallest)
    error_correction=qrcode.constants.ERROR_CORRECT_L, # Standard error correction
    box_size=10, # Pixels per 'box'
    border=4,    # Thickness of the border (minimum is 4)
)

# 2. Add the data
qr.add_data(url)
qr.make(fit=True)

# 3. Create the image (RGB)
img = qr.make_image(fill_color="black", back_color="white")

# 4. Save the file
file_name = "sdu_nauryz_qr.png"
img.save(file_name)

print(f"QR Code successfully generated and saved as {file_name}")
```
