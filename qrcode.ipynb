```python
import qrcode
url = "https://sdu-nauryz2026.website.yandexcloud.kz/"
qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=10,
    border=4,
)
qr.add_data(url)
qr.make(fit=True)
img = qr.make_image(fill_color="black", back_color="white")
file_name = "sdu_nauryz_qr.png"
img.save(file_name)
print(f"QR Code successfully generated and saved as {file_name}")
```
